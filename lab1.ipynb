```python
lab1.loc[:,'faculty_count']=lab1.loc[:,'faculty_count'].str.replace(',', '').astype(float)
lab1.loc[:,'student_faculty_ratio'].astype(float)
lab1.info()
```

```python
lab1['international_students'] = lab1.international_students.str.replace(r'\W+','').astype(float)
```

```python
lab1.info()
```

```python
sns.heatmap(lab1.corr(),annot=True)
plt.title('Correlation Matrix')
```

```python
sns.set_style('white')
sns.lmplot(data=lab1,x='international_students',y='faculty_count',
          aspect=2,height=8)
plt.title('Plot of the linear regression between number of faculty and number of  international students', fontsize=20,
          color="green")
plt.xlabel('Number of International Students')
plt.ylabel('Number of Faculty')
```

```python
sns.set_style('dark')
plt.figure(figsize=(10,12))
sns.barplot(
    data=lab1,x="year", y='international_students', hue='type',palette='viridis')
plt.legend(bbox_to_anchor=(1, 1), loc='upper right')
plt.title('Number of international students in private and public universities from 2017-2022', fontsize=20,
          color="purple")
plt.ylabel('Number of International Students')
plt.xlabel('Type of University')
```

```python
import plotly.graph_objects as go
from plotly.subplots import make_subplots
import plotly.express as px
```

```python
graph=lab1[['international_students','country']].groupby('country').sum().sort_values(by='international_students',ascending=False)
```

```python
sns.set_style('dark')
plt.figure(figsize=(10,12))
sns.barplot(
    data=graph, x='international_students', y=graph.index, palette='viridis')
plt.title('Top 100 Countries of Choice for International Students for the Years 2017-2022', fontsize=15,
          color="purple")
plt.ylabel('Countries', fontsize=20)
plt.xlabel('count', fontsize=20)
```
